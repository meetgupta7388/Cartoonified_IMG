```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Load Image

```python
def read_file(filename):
    img=cv2.imread(filename)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    return img
```

```python
filename="/content/WhatsApp Image 2023-12-14 at 17.56.52_41fc18d8.jpg"
img=read_file(filename)
org_img= np.copy(img)
```

Create Edge mask


INPUT: input image
OUTPUT: edges of image

```python
def edge_mask(img,line_size,blur_value):
    gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    gray_blur=cv2.medianBlur(gray,blur_value)

    edges=cv2.adaptiveThreshold(gray_blur,255,cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY,line_size,blur_value)
    return edges
```

```python
line_size,blur_value=5,5
edges = edge_mask(img,line_size,blur_value)
plt.imshow(edges,cmap="gray")
plt.axis('off')
plt.show()
```

color_quantiztion

```python
def color_quantiztion(img,k):

    # Transform the image
    data=np.float32(img).reshape((-1,3))

    # Determine the criteria
    criteria= (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,20,0.001)

    # Implement K means clustering
    ret, label, center = cv2.kmeans(data,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
    center=np.uint8(center)

    result= center[label.flatten()]
    result= result.reshape(img.shape)

    return result
```

```python
img=color_quantiztion(img,k=8)
plt.imshow(img)
plt.axis('off')
plt.show()
```

Reduce the noise

```python
blurred= cv2.bilateralFilter(img,d=9,sigmaColor=200,sigmaSpace=200)
plt.imshow(blurred)
plt.axis('off')
plt.show()
```

Combine edge_mask with quantized image

```python
def cartoon():
    c=cv2.bitwise_and(blurred,blurred,mask=edges)
    plt.imshow(org_img)
    plt.title("Original Image")
    plt.axis('off')
    plt.show()
    print()
    plt.imshow(c)
    plt.title("Cartoon Image")
    plt.axis('off')
    plt.show()
```

```python
cartoon()
```
